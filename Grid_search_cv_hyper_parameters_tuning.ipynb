# **grid search cv me inear regression me hyperpsrameter tuning**

```python
# Linear Regression grid Search CV hyperparameters 
#{'copy_X': True, 'fit_intercept': True, 'n_jobs': None, 'positive': False}
#linerar regression simple hota hy 2numerics k darmiya

import pandas as pd

import seaborn as sns

df = sns.load_dataset("titanic")

df ['age']= df ['age'].fillna (df ['age'].mean())

X = df [['age']]

y=df['fare']

from sklearn.linear_model import LinearRegression 

from sklearn.model_selection import GridSearchCV #linear regression k sath grid search cv ko import kiya

#create a model
#model aur usky parameter liye

model = LinearRegression()

# define parameter grid

param_grid = {'fit_intercept': [True, False]} #one of this will perform best

#object grid search cv (Creating the model)

grid_search= GridSearchCV(model, param_grid, cv=5, scoring='r2') #gridsearchcv me phly model ko rkh phr hyperparameters ,parameer ki grid me phly true ko check kry ga hr false ko,phr kini dfa esko cross validate krna hy jese 5 dfa,phr ye likhna hy k kis base base pr best ko select krnaq hy, idhr r2 rkhi ku k lineregression me r2 best hoti hy matix k tor pr

#traing the model
grid_search.fit(X,y) #gridsearch ko fit kiya

#print the best parameters

print("Best Parameters: ", grid_search.best_params_)
```

```python
#KNN

# KNN

import pandas as pd

import seaborn as sns

df = sns. load_dataset("titanic")

X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]

y = df['survived']

X= pd.get_dummies (X, columns=['sex'])

X.age.fillna(value = X['age'].mean(), inplace=True)

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import GridSearchCV

import numpy as np

#create a model:

model = KNeighborsClassifier()

# define parameter grid

param_grid = {'n_neighbors': np.arange(1,30,2), 'weights': ['uniform', 'distance']} #hyperparameter 2 hyperparameters liye

#object grid search cv (Creating the model)

#object grid search cv (Creating the model)

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='precision') #precision esliye likha ku k ye classifier hy

#traing the model

grid_search.fit(X,y)

# print the best parameters

print("Best Parameters: ", grid_search.best_params_)

print("Best Score: ", grid_search.best_score_)

```

# **DTC decision tree classifier**

```python
#DTC

# DTC

import pandas as pd

import seaborn as sns

df = sns.load_dataset ("titanic")

y = df ['survived']

X = df [['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
X.age.fillna(value = X['age'].mean(), inplace=True)

X = pd.get_dummies(X, columns=['sex'])

from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV

import numpy as np

#create a model

model = DecisionTreeClassifier()

# define parameter grid

param_grid = { 'max_depth': [3, 5, 7, None], 'min_samples_split': [2, 3, 4]}

#object grid search cv (Creating the model)

grid_search= GridSearchCV (model, param_grid, cv=5, scoring='precision')

#traing the model

grid_search.fit(X,y)


# print the best parameters

print("Best Parameters: ", grid_search.best_params_) 
print("Best Score: ", grid_search.best_score_)
```

# **Random forest**

```python
#RF

#RF

import pandas as pd

import seaborn as sns

df = sns. load_dataset ("titanic")

X = df [['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]

y = df ['survived']

X = pd.get_dummies (X, columns=['sex'])

X.age.fillna(value = X['age'].mean(), inplace=True)

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

import numpy as np

#create a model

model = RandomForestClassifier()

# define parameter grid

param_grid = {'n_estimators': [10, 50, 100],

               'max_depth': [3, 5, 7, None],

                'min_samples_split': [2, 3, 4],

                 'max_features': ['auto', 'sqrt', 'log2']}

#object grid search cv (Creating the model)

grid_search= GridSearchCV (model, param_grid, cv=5, scoring='precision')


#traing the model

grid_search.fit(X,y)

# print the best parameters

print("Best Parameters: ", grid_search.best_params_)

print("Best Score: ", grid_search.best_score_)
```
